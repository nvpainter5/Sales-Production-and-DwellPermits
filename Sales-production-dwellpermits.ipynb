### 2.152 Dwelling permits and sales for 21 European countries

##### (a) Make a scatterplot with sales as the response variable and permits issued for new dwellings as the explanatory variable. Describe the relationship. Are there any outliers or influential observations? 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


plt.rcParams['figure.figsize'] = (8.0, 8.0)
plt.style.use('ggplot')
df  = pd.read_csv('ex02-152meis.csv')
df.plot(kind='scatter',x='DwellPermit',y='Sales',title='Scatterplot - Sales VS Dwellings')
print('The graph below shows weak, positive, and linear relationships between DwellPermit and Sales.\n There are quite a few outliers observed in the graph.')
```

```python
####(b) Find the least-squares regression line and add it to your plot.
```

```python
from scipy.stats import linregress
x = df['DwellPermit']
y = df['Sales']

slope, intercept, r_value, p_value, std_err = linregress(x,y)
plt.plot(x, y, 'o', label='original data')
plt.plot(x, intercept + slope*x, 'r', label='fitted line', color='blue')
plt.title(f"Least - Squares Regression Line {round(intercept, 2)}+{round(slope,2)}*DwellPermit")
plt.xlabel('DwellPermit')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

##### (c) What is the predicted value of sales for a country that has an index of 160 for dwelling permits?

```python
predicted_value = float(intercept) + float(slope)*160
predicted_value
```

##### (d) The Netherlands has an index of 160 for dwelling permits. Find the residual for this country.

```python
actual_value = df.loc[df['Country'] == "Netherlands"]
residual = actual_value['Sales'] - predicted_value
residual
```

##### (e) What percent of the variation in sales is explained by dwelling permits?

```python
print((r_value**2)*100,'%')
```
